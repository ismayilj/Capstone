```python
MasterToronto = pd.read_csv('toronto.csv', delimiter = ',',encoding = "cp1252")
```

```python
MasterToronto = MasterToronto.transpose()
MasterToronto.head()
```

In this cell we select Characteristic row data as our key information about the neighbourhood. As columns in the dataframe left empty, we need to assign this row to the column names, and drop unnecessary row, leaving only neigbourhoods as rows


```python
NewColumnNames = MasterToronto.loc['Characteristic',:]
MasterToronto.rename(columns = NewColumnNames, inplace=True)
```

```python
MasterToronto.drop(['Category', 'Topic', 'Data Source', 'Characteristic',
       'City of Toronto'],inplace=True)
MasterToronto.head()
```

```python
MasterToronto.reset_index(level=0, inplace=True)
MasterToronto.head()
```

```python
MasterToronto.rename(columns={"index":"Hood"},inplace=True)
MasterToronto.head()
```

Further, we need to select the columns we want, as the database is huge. Let's focus on rate of employment as indicator of stability of the area, let's consider I am immigrant wanting to see the situation with Toronto and where it is good to settle down.

Lets select two economic parameters of showing economic well-being in the area, if I want to select a place to settle down and be confident that I will more likely find a job in Toronto.

```python
Torontoanalysis = MasterToronto.loc[:,['Hood','Employment rate']]
Torontoanalysis.head(40)
```

```python
Torontoanalysis['Employment rate'] = pd.to_numeric(Torontoanalysis['Employment rate'], errors='ignore', downcast='float')
Torontoanalysis.dtypes
```

```python
Torontoanalysis.values[:,1:]
```

Meanwhile let's make a cluste analysis, I would like to share the neighbourhoods by 3 cluster of employment rate.

```python
cluster_dataset = Torontoanalysis.values[:,1:]
```

```python
num_clusters = 3

k_means = KMeans(init="k-means++", n_clusters=num_clusters, n_init=12)
k_means.fit(cluster_dataset)
labels = k_means.labels_
Torontoanalysis['Labels']= labels

print(labels)
Torontoanalysis.groupby('Labels').mean()
```

As a result I have got 3 clusters of Toronto neighbourhoods by employment rates.
